#서울 에 있는 대학교의 위치를 마커로 표시

```python
df = pd.read_excel('./서울지역 대학교 위치.xlsx', engine='openpyxl') 
```

```python
seoul_map2 = folium.Map(location=[37.55,126.98], tiles='Stamen Terrain', zoom_start=15)  
```

```python
for name,lat, lng in zip (df.index, df.위도, df.경도):
    folium.Marker([lat, lng], popup=name).add_to(seoul_map2)
```

```python
seoul_map2
```

```python
seoul_map2.save('./seoul2.html')
```

circle marker

```python
for name,lat, lng in zip (df.index, df.위도, df.경도):
    folium.CircleMarker([lat, lng], radius=10, color='brown', fill=True, fill_color='coral', fill_opacity=0.7, popup=name ).add_to(seoul_map2)
seoul_map2.save('./seoul_circle.html')
```

단계구분도

```python
import json
```

```python
file_path ='/content/경기도인구데이터.xlsx'
df = pd.read_excel (file_path, index_col='구분', engine='openpyxl')
```

```python
df
```

```python
df.columns = df.columns.map(str)
```

```python
geo_path ='./경기도행정구역경계.json'
try:
    geo_data = json.load(open(geo_path, encoding = 'utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding = 'utf-8-sig')) 

g_map = folium.Map(location =[37.5502, 126.982], tiles='Stamen Terrain', zoom_start=9)
```

```python
#출력할 연도 선택 (2007 ~ 2017)년 중 선택
year ='2017'
folium.Choropleth(geo_data=geo_data, data= df[year], columns = [df.index, df[year]],fill_color= 'YlOrRd', fill_opacity=0.7, 
                  line_opacity=0.3,threshold_scale=[10000,100000,300000,500000,700000],key_on='feature.properties.name').add_to(g_map)
```

```python
g_map
```

```python
g_map.save('./경기인구' + year + '.html')
```
