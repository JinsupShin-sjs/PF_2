```python
import pandas as pd
from urllib.parse import quote
```

```python
site ='https://www.siksinhot.com/ko/'
url=f'{site}/search?keywords={quote("신논현역")}'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

필요한 데이터 찾기

```python
lis=soup.find_all('li')
len(lis)
```

```python
lis = soup.select_one('.localFood_list').find_all('li')
len(lis)
```

```python
lis = soup.select('ul.localFood_list >  li')
len(lis)
```

```python
li= lis[1]
li
```

```python
store=li.select_one('.textBox > h2').get_text()
store
```

```python
#float, int 전환 시 반드시 예외 처리하 것 (일부 없는 자료 가 있어서 / 점수가 없을 수 있어서 예외처리 해야함)
```

```python
line =[]
for li in li.select_one('.textBox > h2').get_text():
    try:
        score = float(li.select_one('.score').get_text())
    except:
        score =0.0
    place =li.select_one('.cate').get_text().split('/')[0]
    menu =li.select_one('.cate').get_text().split('/')[1]
    line.append({'상호':store,'평점':score, '주소':place, '메뉴':menu })
```
