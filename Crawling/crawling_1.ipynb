```python
from bs4 import BeautifulSoup       
```

```python
with open('00.Example.html')as f:
    soup = BeautifulSoup(f, 'html.parser')
```

1. find(), find_all()

```python
#첫번쩨 div 태그 내용 가져오기
first_div = soup.find('div')
print(first_div)
```

```python
#해당 되는 테그를 모두 가져오기
ps = first_div.find_all('p')
print (ps)
```

```python
for ptag in ps:
    print(ptag)
```
