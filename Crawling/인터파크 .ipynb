```python
trans = '-'
for li in sub_lis:
    text = li.get_text().strip()
# print(text)
    if '발행:'in text:
        date = text.split(':')[1].strip()
        date = re.sub('[^0-9', '',date)
    elif '쪽수:'in text:
        page = int(text.split(':')[1].strip())
    elif '역:'in text:
        trans = text.split(':')[1].strip()
    elif '저:'in text:
        author = text.split(':')[1].strip()
print(date, page, trans, author)
```
