```python
#https://www.melon.com/chart/     순위 / 제목 / 가수 / 제목

import requests
from bs4 import BeautifulSoup
url = 'https://www.melon.com/chart/'
header ={'User-Agent':'Mozila/5.0 (Windows NT 10.0: Win64: x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.70 Safari/537.36'}
req = requests.get(url, headers=header)
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
import pandas as pd
```

```python
soup
```

```python
trs = soup.find_all('tr')
len(trs) 
```

```python
trs=soup.select('#frm > div > table > tbody > tr')
len(trs)
```

```python
tr= trs[0]
tr
```

```python
rank = int(tr.select_one('.rank').get_text().strip())
rank
```

```python
title = tr.select_one('.ellipsis.rank01').get_text().strip()
title
```

```python
artist = tr.select_one('.ellipsis.rank02').get_text().strip()
artist
```

```python
album = tr.select_one('.ellipsis.rank03').get_text().strip()
album
```

```python
# like는 selenium을 활요해야함
# like = tr.select_one('.cnt').get_text().strip()
# like
```

```python
line =[]
for tr in trs:
    rank = int(tr.select_one('.rank').get_text().strip())
    title = tr.select_one('.ellipsis.rank01').get_text().strip()
    artist = tr.select_one('.ellipsis.rank02').get_text().strip()
    album = tr.select_one('.ellipsis.rank03').get_text().strip()
    line.append([rank, title, artist, album])
```

```python
df = pd.DataFrame(line, columns=['순위', '제목', '가수', '앨범'])
df.head()
```
